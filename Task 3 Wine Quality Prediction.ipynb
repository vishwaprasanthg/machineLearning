### GUI

```python
from tkinter import *
from sklearn.preprocessing import StandardScaler
import joblib
```

```python
from tkinter import *
import joblib
import numpy as np


# Function to make prediction
def show_entry_fields():
    try:
        p1 = float(e1.get())
        p2 = float(e2.get())
        p3 = float(e3.get())
        p4 = float(e4.get())
        p5 = float(e5.get())
        p6 = float(e6.get())
        p7 = float(e7.get())
        p8 = float(e8.get())
        p9 = float(e9.get())
        p10 = float(e10.get())
        p11 = float(e11.get())

        # Load the model, scaler, and PCA
        model = joblib.load("wine_quality_prediction")
        result = model.predict(pca.transform(st.transform([[p1, p2, p3, p4, p5, p6, p7, p8, p9, p10, p11]])))

        # Display result and print to console
        if result[0] == 0:
            result_label.config(text="Bad Quality Wine", fg="red")
            print("Bad Quality Wine")
        else:
            result_label.config(text="Good Quality Wine", fg="green")
            print("Good Quality Wine")

    except ValueError:
        result_label.config(text="Please enter all values correctly", fg="orange")


# Function to load sample data
def load_sample_data():
    sample_data = [7.3, 0.65, 0.0, 1.2, 0.065, 15.0, 21.0, 0.9946, 3.39, 0.47, 10.0]
    entries = [e1, e2, e3, e4, e5, e6, e7, e8, e9, e10, e11]
    for i, entry in enumerate(entries):
        entry.delete(0, END)
        entry.insert(0, str(sample_data[i]))


# Function to reset fields
def reset_fields():
    entries = [e1, e2, e3, e4, e5, e6, e7, e8, e9, e10, e11]
    for entry in entries:
        entry.delete(0, END)
    result_label.config(text="")


# Create main window
master = Tk()
master.title("Wine Quality Prediction Using Machine Learning")
master.geometry("400x550")
master.configure(bg="#2C3E50")

# Title Label
Label(
    master,
    text="Wine Quality Prediction Using ML",
    bg="#34495E",
    fg="white",
    font=("Arial", 16, "bold"),
).pack(fill=X, pady=20)

input_frame = Frame(master, bg="#2C3E50")
input_frame.pack(pady=10)

labels = [
    "Fixed Acidity",
    "Volatile Acidity",
    "Citric Acid",
    "Residual Sugar",
    "Chlorides",
    "Free Sulfur Dioxide",
    "Total Sulfur Dioxide",
    "Density",
    "pH",
    "Sulphates",
    "Alcohol",
]
entries = []

for i, label in enumerate(labels):
    Label(input_frame, text=label, bg="#2C3E50", fg="white", font=("Arial", 10)).grid(
        row=i, column=0, pady=5, padx=10, sticky=W
    )
    entry = Entry(input_frame)
    entry.grid(row=i, column=1, pady=5, padx=10)
    entries.append(entry)

e1, e2, e3, e4, e5, e6, e7, e8, e9, e10, e11 = entries

button_frame = Frame(master, bg="#2C3E50")
button_frame.pack(pady=20)

Button(
    button_frame,
    text="Predict",
    command=show_entry_fields,
    bg="#1ABC9C",
    fg="white",
    font=("Arial", 10, "bold"),
    width=10,
).grid(row=0, column=0, padx=10)

Button(
    button_frame,
    text="Load Sample",
    command=load_sample_data,
    bg="#3498DB",
    fg="white",
    font=("Arial", 10, "bold"),
    width=10,
).grid(row=0, column=1, padx=10)

Button(
    button_frame,
    text="Reset",
    command=reset_fields,
    bg="#E74C3C",
    fg="white",
    font=("Arial", 10, "bold"),
    width=10,
).grid(row=0, column=2, padx=10)

result_label = Label(master, text="", font=("Arial", 12, "bold"), bg="#2C3E50")
result_label.pack(pady=10)

mainloop()
```
